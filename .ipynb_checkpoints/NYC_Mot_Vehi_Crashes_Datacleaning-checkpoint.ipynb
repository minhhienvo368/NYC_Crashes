We are going to use a dataset about New York City Motor Vehicles Crashes taken from NYC Open Data which is a great resource for data sets. This dataset is quite small and this notebook will just be an overview of different stuff to look out for when cleaning data.

https://data.cityofnewyork.us/Public-Safety/Motor-Vehicle-Collisions-Crashes/h9gi-nx95

```python
# Drop the redundant columns: e.g. `location`
df = df.drop(columns= 'location', axis=1)
df.columns
```

```python
# Change dtypes
import datetime

#datetime.datetime.strptime("2013-07-10T11:00:00.000Z","%Y-%m-%dT%H:%M:%S.%fZ")
#new_format = "%d-%m-%Y"
#df['crash_date'] = df['crash_date'].datetime.strftime(new_format)

df['crash_date'] = pd.to_datetime(df['crash_date'])
df.crash_date.dtypes
```

### Feature slicing

```python
df["day"] = df['crash_date'].map(lambda x: x.day)
df["month"] = df['crash_date'].map(lambda x: x.month)
df["year"] = df['crash_date'].map(lambda x: x.year)

# Split the feature Date into day, month and year
df[['crash_date', 'day', 'month', 'year']].head()
```

```python
df['crash_time'] = df['crash_time'].apply(lambda x: pd.Timestamp(x))

#df['crash_timestamp'] = df.crash_time.values.astype(np.int64) // 10**9
```

```python
df['crash_timestamp'].unique()
```

### Convert dtypes from float to integer

```python
#df['zip_code'] = df['zip_code'].astype('int64')
df['zip_code'] = df['zip_code'].dropna().apply(np.int64)
```

```python
cols_float2int = ['zip_code']
for col in cols_float2int:
    df[col] = df[col].apply(lambda x: int(x) if x == x else "")
```
